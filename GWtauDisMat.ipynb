```python
# basic inports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
Recovery = pd.read_excel('GWtauFRAP.xlsx')
#remove column labels and replace with "lead" for first 9 and "lag" for last 10
Recovery.columns = ['lead', 'lead', 'lead', 'lead', 'lead', 'lead', 'lead', 'lead', 'lead', 'lead', 'lag', 'lag', 'lag', 'lag', 'lag', 'lag', 'lag', 'lag', 'lag', 'lag']
Recovery.head()
```

```python
Recruitment = pd.read_excel('GWtauRecruitment.xlsx')
Recruitment.columns = ['lead', 'lead', 'lead', 'lead', 'lead', 'lead', 'lead', 'lead', 'lead', 'lead', 'lag', 'lag', 'lag', 'lag', 'lag', 'lag', 'lag', 'lag', 'lag', 'lag']
Recruitment.head()
```

# Recruitment

```python
# heatmap of values in Recruitment
plt.figure(figsize=(20,10))
sns.heatmap(Recruitment, annot=True, cmap='viridis')
plt.title('Recruitment')
plt.show()
```

```python
# perform clustering given the distance matrix Recruitment1
from scipy.cluster import hierarchy
from scipy.spatial import distance

# convert the redundant n*n square matrix form into a condensed nC2 array
# distArray[{n choose 2}-{n-i choose 2} + (j-i-1)] is the distance between points i and j
distArray = distance.squareform(Recruitment)
# calculate the linkage matrix
Z = hierarchy.linkage(distArray, 'single')
# plot dendrogram
plt.figure(figsize=(20,10))
dn = hierarchy.dendrogram(Z)
plt.title('Recruitment')
plt.show()
```

# Recovery

```python
# heatmap of values in Recruitment
plt.figure(figsize=(20,10))
sns.heatmap(Recovery, annot=True)
plt.title('Recovery')
plt.show()
```

```python
# perform clustering given the distance matrix Recovery1
from scipy.cluster import hierarchy
from scipy.spatial import distance

# convert the redundant n*n square matrix form into a condensed nC2 array
# distArray[{n choose 2}-{n-i choose 2} + (j-i-1)] is the distance between points i and j
distArray = distance.squareform(Recovery)
# calculate the linkage matrix
Z = hierarchy.linkage(distArray, 'single','euclidean',True)
# plot dendrogram
plt.figure(figsize=(20,10))
dn = hierarchy.dendrogram(Z)
plt.title('Recovery')
plt.show()
```
